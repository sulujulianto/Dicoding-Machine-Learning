```python
from google_play_scraper import reviews, Sort
import pandas as pd

# Ambil data ulasan Tokopedia dari Google Play
app_package = 'com.tokopedia.tkpd'  # ID aplikasi Tokopedia
all_reviews = []

# Loop untuk ambil data bertahap (tiap panggilan max 200 ulasan)
for _ in range(20):  # 20 * 200 = 4000 ulasan
    result, _ = reviews(
        app_package,
        lang='id',             # Bahasa Indonesia
        country='id',          # Indonesia
        sort=Sort.NEWEST,      # Ambil dari yang terbaru
        count=200,             # Jumlah per batch
        filter_score_with=None # Ambil semua rating (1‚Äì5)
    )
    all_reviews.extend(result)

# Simpan ke DataFrame
df = pd.DataFrame(all_reviews)
df = df[['userName', 'score', 'content']]

# Ganti nama kolom biar rapi
df.columns = ['nama_pengguna', 'rating', 'ulasan']

# Simpan ke file CSV
df.to_csv('dataset.csv', index=False, encoding='utf-8')
print(f"Total data terkumpul: {len(df)} ulasan")
df.head()
```
